#ML

#1.Split data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn import metrics

x=df.drop('Attrition',axis=1)
y=df['Attrition']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2)
```

##LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr1=lr.fit(x_train,y_train)
ypred1=lr1.predict(x_test)
score = lr1.score(x_train, y_train)
score
```

```python
score_lr=accuracy_score(y_test,ypred1)
score_lr
```

```python

confusion_matrix(y_test,ypred1)

```

```python
precision_score(y_test,ypred1)
```

```python
recall_score(y_test,ypred1)
```

```python
f1_score(y_test,ypred1)
```

```python
classification_report_lr = classification_report(y_test, ypred1)
print(classification_report_lr)
```

##SVM

```python
from sklearn import svm
sv=svm.SVC()
sv1=sv.fit(x_train,y_train)
ypred3=sv1.predict(x_test)
score_sv = sv.score(x_train, y_train)
score_sv
```

```python

score_svm=accuracy_score(y_test,ypred3)
score_svm
```

```python
classification_report_sv = classification_report(y_test, ypred3)
print(classification_report_sv)
```

```python
from sklearn.metrics import roc_auc_score

micro_roc_auc_ovr = roc_auc_score(
    y_test,ypred3,
)
```

```python
from sklearn.metrics import roc_curve
fprr, tprr, thresholdsr = roc_curve( y_test,ypred3)
plt.figure(figsize=(8, 6))
plt.plot(fprr, tprr, linewidth=2)
plt.plot([0, 1], [0, 1], 'r--')
plt.grid(True)
plt.show()
```

#using pipeline

```python
# Defining a pipeline for Polynomial Kernel SVM classifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="poly", degree=10, coef0=5, C=10))# coef0 adding to dot product which try to find relation
])


svm=svm_clf.fit(x_train, y_train)
ypred4=knn1.predict(x_test)
```
